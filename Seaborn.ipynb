```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# seaborn : statistical data visualization
# It contains somedatasets also.
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset("tips")
```

```python
df
```

```python
# Numerical data plotting

sns.relplot( x = 'total_bill', y = 'tip', data = df)
# it helps us to understand the relationship between two variables.
# by default relplot gives scatterplot.
# if we give kind = 'line' then it will give line chart.
```

```python
# relplot
# This function provides access to several different axes-level functions
# that show the relationship between two variables with semantic mappings of subsets.
# The kind parameter selects the underlying axes-level function to use:

# scatterplot() (with kind="scatter"; the default)
# lineplot() (with kind="line")
```

```python
sns.scatterplot(x = df.total_bill, y = df.tip, data = df, hue = 'smoker')
```

```python
# as the bill increase, the tip also increases
# highest tip was paid by a smoker
```

```python
sns.scatterplot(x = df.total_bill, y = df.tip, data = df, hue = 'size')
```

```python
# hue gives different colors.
# style gives different symbols.
```

```python
sns.scatterplot(x = df.total_bill, y = df.tip, data = df, style = 'smoker', hue = 'time')
```

```python
sns.scatterplot(x = df.total_bill, y = df.tip, data = df, style = 'size')
```

```python
# relplot to lineplot <<< we generated a data
value = np.random.randn(500).cumsum()
time = np.arange(500)

df = pd.DataFrame({'time': time, 'value': value})
```

```python
df
```

```python
sns.relplot(x = 'time', y = 'value', kind = 'line', data = df, sort = True)
```

```python
fmri = sns.load_dataset('fmri')
```

```python
fmri
```

```python
sns.relplot(x = 'timepoint',y = 'signal' , kind = 'line', data = fmri) # shadow is the Confidence interval
```

```python
fmri.info()
```

```python
sns.relplot(x = 'timepoint',y = 'signal' , kind = 'line', data = fmri, ci = 'sd') # confidence interval = standard deviation
```

```python
sns.relplot(x = 'timepoint',y = 'signal' , estimator = None, kind = 'line', data = fmri) # without confidence interval.
```

```python
sns.relplot(x = 'timepoint',y = 'signal' , kind = 'line', data = fmri, hue = 'event')
```

```python
sns.relplot(x = 'timepoint',y = 'signal' , kind = 'line', data = fmri, style = 'region',hue = 'event')
```

```python
sns.relplot(x = 'timepoint',y = 'signal' , kind = 'line', data = fmri, style = 'region',hue = 'event', markers = True)
```
